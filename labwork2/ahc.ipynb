```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X = [1, 2, 9, 12, 20]
X_d = [[i] for i in X]
```

```python
complete_linkage = linkage(X_d, "complete")
single_linkage = linkage(X_d, "single")
```

```python
dendrogram(complete_linkage, labels=X)
```

```python
dendrogram(single_linkage, labels=X)
```

```python
data = pd.read_csv("./iris/iris.data", header=None)
feature_column = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
data.columns = [*feature_column, "class"]
features_data = data[feature_column]
```

```python
s = linkage(features_data, "single")
c = linkage(features_data, "complete")
```

```python
plt.figure(figsize=(150,60))
dendro = dendrogram(c)
img = plt.savefig("./iris_complete.png", format="png")
```

```python
plt.figure(figsize=(150,60))
dendro = dendrogram(s)
img = plt.savefig("./iris_single.png", format="png")
```

```python
data = pd.read_csv("./Rice_Cammeo_Osmancik.csv", header=None)
feature_column = ['area', 'perimeter', 'compactness', 'length', 'width', 'asymmetry', 'groove_length']
data.columns = [*feature_column, "class"]
features_data = data[feature_column]
```

```python
s = linkage(features_data, "single")
c = linkage(features_data, "complete")
```

```python
plt.figure(figsize=(30,8))
dendro = dendrogram(s)
img = plt.savefig("./rice_single.png", format="png")
```

```python
plt.figure(figsize=(30,8))
dendro = dendrogram(c)
img = plt.savefig("./rice_complete.png", format="png")
```
